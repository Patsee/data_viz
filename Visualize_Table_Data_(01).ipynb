```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

### Parallel Coordinates

```
pd.plotting.parallel_coordinates(df ,'Name',)
```

### ปรับ color

```
pd.plotting.parallel_coordinates(df ,'Name', color=('#F50909','#6EFF33','#0000FF'))
```

```
pd.plotting.parallel_coordinates(df ,'Name', color=('R','G','m')) #สามารถย่อได้  R G B C M Y K W
```

```
pd.plotting.parallel_coordinates(df ,'Name', color=((1,0,0), (0,1,0), (0,0,1))) #rgb value(r,g,b) ->[0,1]
```

## เริ่ม matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas เป็น numpy array

```
g = df.iloc[:,:-1].to_numpy()
g
```

## Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15]   #กำหนดขนาดของ figure
plt.imshow(g[:10,:],interpolation='nearest')
```

```
plt.imshow(g[50:100,:],interpolation='nearest')
```

การใช้ subplot

```
plt.subplot(1,3,1) #subplot (มีกี่แถว ,มี่กี่หลัก ,เป็นตัวที่เท่าไหร่)
plt.imshow(g[:50,:],interpolation='nearest') 

plt.subplot(1,3,2)
plt.imshow(g[50:100,:],interpolation='nearest')

plt.subplot(1,3,3)
plt.imshow(g[100:150,:],interpolation='nearest')
```
