```r
datcorr = cor(data1) 
datKMO =KMO(datcorr) 
n=120
datBart=cortest.bartlett(datcorr,n) 
print(datBart$p.value,4) 
```

```r
datEig = eigen(datcorr) 
datEigmean = mean(datEig$values) 
numfac = sum(datEig$values>=1)
print(numfac)
```

```r
 data1Fa = principal(datcorr,nfactors=numfac,rotate="none") # กรณีไม่หมุนแกน
 print(data1Fa)
```

```r
X = data1
library(psych)
fa <- principal(X, nfactors = numfac, rotate = "none", covar = FALSE)
round(fa$values,3)
```

```r
plot(fa$values, type = "b", pch = 19, col="blue", main = "Scree Plot",xlab = "Eigenvalue number",ylab = "Eigenvalue size")
```
